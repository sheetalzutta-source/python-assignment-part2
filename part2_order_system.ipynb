```python
# Task 2 - Cart Operations

# Simulate a customer placing an order. Represent the cart as a list of dictionaries, starting empty:
cart = []
# Each cart entry should look like: {"item": "Paneer Tikka", "quantity": 2, "price": 180.0}

# 1. Write logic to add an item to the cart:
# 1.1 Check if the item exists in menu.
# 1.2 Check if the item is marked as available.
# 1.3 If the item is already in the cart, increase its quantity instead of adding a duplicate entry.
# 1.4 If the item doesn't exist in the menu or is unavailable, print an appropriate message and do not modify the cart.

def add_to_cart(item_name, quantity):
    if item_name not in menu:
        print(f"Error: '{item_name}' does not exist in the menu.")
        return
    if not menu[item_name]["available"]:
        print(f"Sorry, '{item_name}' is currently unavailable.")
        return
    
    price = menu[item_name]["price"]
    
    # Check if item is already in cart
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] += quantity
            print(f"Updated '{item_name}' quantity to {entry['quantity']} in the cart.")
            return
    
    # If item is not in cart, add new entry
    cart.append({"item": item_name, "quantity": quantity, "price": price})
    print(f"Added '{item_name}' (x{quantity}) to the cart at ₹{price:.2f} each.")

# 2. Write logic to remove an item from the cart by name. If the item is not in the cart, print a message.
def remove_from_cart(item_name):
    for entry in cart:
        if entry["item"] == item_name:
            cart.remove(entry)
            print(f"Removed '{item_name}' from the cart.")
            return
    print(f"Error: '{item_name}' is not in the cart.")


# 3. Write logic to update the quantity of an item already in the cart.
def update_cart(item_name, new_quantity):
    for entry in cart:
        if entry["item"] == item_name:
            if new_quantity <= 0:
                cart.remove(entry)
                print(f"Removed '{item_name}' from the cart due to zero or negative quantity.")
            else:
                entry["quantity"] = new_quantity
                print(f"Updated '{item_name}' quantity to {new_quantity} in the cart.")
            return
    print(f"Error: '{item_name}' is not in the cart.")

#4. Using your logic above, simulate the following sequence. Print the cart state after each step:

# 4.1 Add "Paneer Tikka" × 2 and print the cart
add_to_cart("Paneer Tikka", 2)
print("Current cart:", cart)
print () # add a blank line for better readability

# 4.2 Add "Gulab Jamun" × 1
add_to_cart("Gulab Jamun", 1)
print("Current cart:", cart)
print () # add a blank line for better readability

# 4.3 Add "Paneer Tikka" × 1 → quantity should update to 3, not create a new entry
add_to_cart("Paneer Tikka", 1)
print("Current cart:", cart)
print () # add a blank line for better readability

# 4.4 Try to add "Mystery Burger" → does not exist in menu
add_to_cart("Mystery Burger", 1)
print("Current cart:", cart)
print () # add a blank line for better readability

# 4.5 Try to add "Chicken Wings" → exists but is unavailable
add_to_cart("Chicken Wings", 1)
print("Current cart:", cart)
print () # add a blank line for better readability

# 4.6 Remove "Gulab Jamun"
remove_from_cart("Gulab Jamun")
print("Current cart:", cart)
print () # add a blank line for better readability

# 5 Print a final Order Summary:

# ========== Order Summary ==========
# Paneer Tikka       x3    ₹540.00
# ...
# ------------------------------------
# Subtotal:                ₹XXX.00
# GST (5%):                ₹ XX.XX
# Total Payable:           ₹XXX.XX
# ====================================
def print_order_summary(cart):
    print("========== Order Summary ==========")
    subtotal = 0.0
    for entry in cart:
        item_total = entry["quantity"] * entry["price"]
        subtotal += item_total
        print(f"{entry['item']:<20} x{entry['quantity']:<5} ₹{item_total:>8.2f}")
    print("-" * 40)
    gst = subtotal * 0.05
    total_payable = subtotal + gst
    print(f"{'Subtotal:':<25} ₹{subtotal:>8.2f}")
    print(f"{'GST (5%):':<25} ₹{gst:>8.2f}")
    print(f"{'Total Payable:':<25} ₹{total_payable:>8.2f}")
    print("====================================")
print_order_summary(cart)

```

```python
# Task 3 - Inventory Tracker with Deep Copy. Use the provided inventory dictionary.

# 1. Deep copy inventory into a variable called inventory_backup before making any changes.
# To demonstrate this works, manually change one stock value in inventory and print both inventory and inventory_backup to show that the backup is unaffected. Then restore inventory to its original state before continuing.
# (Hint: import copy and use copy.deepcopy().)

import copy
inventory_backup = copy.deepcopy(inventory)

# Manually change one stock value in inventory
inventory["Paneer Tikka"]["stock"] = 5
print("Current inventory:", inventory)
print("\nInventory backup:", inventory_backup)

# Restore inventory to original state
inventory = copy.deepcopy(inventory_backup)
print("\nRestored inventory:", inventory)

# 2. Simulate order fulfilment: deduct the quantities from the final cart in Task 2 from the corresponding items in inventory. 
# If stock for an item is insufficient, print a warning and deduct only what is available (do not go below 0).
for entry in cart:
    item_name = entry["item"]
    quantity_ordered = entry["quantity"]
    
    if item_name in inventory:
        available_stock = inventory[item_name]["stock"]
        if available_stock >= quantity_ordered:
            inventory[item_name]["stock"] -= quantity_ordered
            print(f"\nFulfilled order for '{item_name}' (x{quantity_ordered}). Remaining stock: {inventory[item_name]['stock']}")
        else:
            inventory[item_name]["stock"] = 0
            print(f"\nWarning: Only {available_stock} '{item_name}' available. Fulfilled partial order. Remaining stock: 0")
    else:
        print(f"Error: '{item_name}' not found in inventory.")

# 3. After deduction, loop through inventory and print a Reorder Alert for every item whose stock is at or below its reorder_level:
for item_name, details in inventory.items():
    if details["stock"] <= details["reorder_level"]:
        print(f"\n⚠ Reorder Alert: {item_name} — Only {details['stock']} unit(s) left (reorder level: {details['reorder_level']})")

# 4. Print both inventory and inventory_backup at the end to confirm they differ — proving the deep copy protected the original.
print("\nFinal inventory after order fulfilment:", inventory)
print("\nInventory backup (should be unchanged):", inventory_backup)

```
